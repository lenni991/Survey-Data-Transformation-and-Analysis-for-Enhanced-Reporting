```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import numpy as np
```

```python
df = pd.read_excel("Edited Data - Survey Monkey Output.xlsx", sheet_name="Edited_data")
```

```python
df.head(10)
```

```python
df_modified = df.copy()
```

```python
df_modified.info()
```

```python
#droping columns we will not need
df_modified.drop(columns=['Start Date','End Date','First Name','Last Name', 'Email Address', 'Custom Data 1'],inplace=True)
df_modified.info()
```

```python
#organizing lists to be put into the melt method

#df_modified.melt()

# it takes +1 args, usually id_vars and value_vars
# id_vars -> Columns to keep (stay fixed, e.g., IDs)
# value_vars -> Columns to unpivot (convert from wide to long)
# we only need to set id_vars, and the method is going to unpovit all other columns by default when it's not specified
```

```python
id_vars = list(df_modified.columns)[0:8]
```

```python
df_melted = df_modified.melt(id_vars=id_vars, var_name='Question + SubQuestion',value_name='Answer')
df_melted
```

```python
df_melted[df_melted['Respondent ID']==6638341389]
```

```python
questions_import = pd.read_excel("Edited Data - Survey Monkey Output.xlsx", sheet_name='Question')
```

```python
questions_import
```

```python
questions = questions_import.copy()
```

```python
questions.columns
```

```python
questions.drop(columns=['Raw Question','Raw Subquestions','Subquestion','Unnamed: 5'], inplace=True)
questions
```

```python
dataset_merged = pd.merge(how='left', left=df_melted, right=questions, left_on='Question + SubQuestion', right_on='Question + SubQuestion')
```

```python
dataset_merged
```

```python
df_melted.shape, dataset_merged.shape
```

```python
# how many unique respondents answered each question, ignoring missing answers.

respondents = dataset_merged[dataset_merged['Answer'].notnull()].groupby('Question')['Respondent ID'].nunique()

respondents.reset_index()
```

```python
respondents = respondents.to_frame().rename(columns={'Respondent ID': 'Respondents'})
respondents
```

```python
dataset_merged_two = pd.merge(how='left', left=dataset_merged, right=respondents, left_on='Question', right_on='Question')
dataset_merged.shape, dataset_merged_two.shape
```

```python
dataset_merged_two
```

```python
#how many people answered the same answer per question
same_answer = dataset_merged[dataset_merged['Answer'].notnull()].groupby(['Question + SubQuestion', 'Answer'])['Respondent ID'].nunique()
```

```python
same_answer.to_frame()
```

```python
same_answer = same_answer.to_frame().rename(columns={'Respondent ID': 'same_answer'})
same_answer
```

```python
dataset_merged_three = pd.merge(how='left', left=dataset_merged_two, right=same_answer, left_on=['Question + SubQuestion','Answer'], right_on=['Question + SubQuestion','Answer'])
dataset_merged_two.shape, dataset_merged_three.shape
```

```python
dataset_merged_three
```

```python
dataset_merged_three['same_answer'] = dataset_merged_three['same_answer'].fillna(0)
dataset_merged_three
```

```python
dataset_merged_three.columns[1:8]
```

```python
output = dataset_merged_three.copy()
columns_names = {
    'Identify which division you work in.-Response': 'Division Primary',
    'Identify which division you work in.-Other (please specify)': 'Divison Secondary',
    'Which of the following best describes your position level?-Response' : 'Position',
    'Which generation are you apart of?-Response' : 'Generation',
    'Please select the gender in which you identify.-Response' : 'Gender',
    'Which duration range best aligns with your tenure at your company?-Response': 'Tenure',
    'Which of the following best describes your employment type?-Response': 'Employee Type'
    }

output.rename(columns=columns_names, inplace=True)
```
